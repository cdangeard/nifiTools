```python
from typing import List

def maxEnvelopes(envelopes: List[List[int]]) -> int:
    def canFit(son : List[int], dad : List[int]):
        return son[0] < dad[0] and son[1] < dad[1]
    
    def getBigestCombinaison(doll : List[int], envelopes : List[List[int]],
                                comb : List[List[int]]):
        sons = [son for son in envelopes if canFit(son,doll)]
        print(sons)
        if len(sons) <= 0:
            return comb
        for son in sons:
            tcomb = getBigestCombinaison(son, sons, comb.append(son))
            if len(tcomb) > len(comb):
                comb = tcomb
        return comb

    fakeBigDoll = [10**5,10**5]
    return len(getBigestCombinaison(fakeBigDoll, envelopes, [])) - 1

maxEnvelopes([[5,4],[6,4],[6,7],[2,3]])
```
